```python
#1 
#Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python.
#Matplotlib is used for creating visualizations and plots in Python.It provides a wide range of plotting functions and styles for representing data.
# Plots that can be used are 1.scatterplot 2.lineplot 3.histogram 4.barplot 5.piechart
```

```python


import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
#2
import matplotlib.pyplot as plt
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Assignment')
plt.scatter(x,y)
plt.show()
```

```python
#3 The subplot() function in Matplotlib is used to create multiple plots within the same figure. 
#It allows you to arrange plots in a grid-like fashion, specifying the number of rows and columns and the position of each subplot.

# Data for line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Data for line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Data for line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Data for line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Creating subplots
plt.figure(figsize=(10, 8))

# Plot 1
plt.subplot(2, 2, 1)
plt.plot(x1, y1)
plt.title('Line Plot 1')

# Plot 2
plt.subplot(2, 2, 2)
plt.plot(x2, y2)
plt.title('Line Plot 2')

# Plot 3
plt.subplot(2, 2, 3)
plt.plot(x3, y3)
plt.title('Line Plot 3')

# Plot 4
plt.subplot(2, 2, 4)
plt.plot(x4, y4)
plt.title('Line Plot 4')

plt.tight_layout()
plt.show()
```

```python
#4 A bar plot is a type of visualization that presents categorical data with rectangular bars. 
#The length or height of each bar represents the magnitude of the data being plotted. 
#Bar plots are used to compare the quantities of different categories or to show the distribution of categorical data.

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Creating a vertical bar plot
plt.figure(figsize=(8, 6))
plt.bar(company, profit, color='skyblue')
plt.xlabel('Company')
plt.ylabel('Profit ($)')
plt.title('Profits of Companies')
plt.show()
```

```python
#5

#A box plot is a graphical representation of the distribution of a dataset based on five summary statistics: minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum

# it is used for 1.Comparing Distributions ,2.Identifying Outliers,3.Displaying Variability 


# Generating random data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Creating a box plot
plt.figure(figsize=(8, 6))
plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'])
plt.xlabel('Box')
plt.ylabel('Values')
plt.title('Box Plot of Box 1 and Box 2')
plt.show()
```
