```python
#2 

def reindex_dataframe(df):
    # Reset the index to the default integer index
    df = df.reset_index(drop=True)
    
    # Create a new index starting from 1 and incrementing by 2
    new_index = pd.Series(range(1, len(df) * 2, 2))
    
    # Set the new index to the DataFrame
    df.index = new_index
    
    return df

# Example DataFrame
df = pd.DataFrame({
    'A': [10, 20, 30, 40],
    'B': [100, 200, 300, 400],
    'C': [1000, 2000, 3000, 4000]
})

# Re-index the DataFrame
new_df = reindex_dataframe(df)

# Print the new DataFrame
print(new_df)
```

```python
#8 
 
    
df3 = pd.DataFrame({"A":[3,2,7,9,4],
                    "B":[6,7,8,4,8],
                    "C":[78,23,45,6,7]})

def select_rows(df):
    selected_row = df[(df['A']>5) & (df['B']<10)]
    return selected_row

select_rows(df3)
```

```python
#9 
df4 = pd.DataFrame({'Values':[20,365,2789,789,432,43]})

def central_tendency(df):
    mean = df.mean()
    median = df.median()
    std = df.std()
    print (f"The values are {mean} ,{median},{std}")

central_tendency(df4)
```

```python
#10

import pandas as pd
data = {
    'Date': pd.date_range(start='2024-01-01', periods=10),
    'Sales': [100, 120, 110, 130, 140, 150, 160, 170, 180, 190]
}
df5 = pd.DataFrame(data)
def calculate_moving_average(df):
    df['MovingAverage'] = df['Sales'].rolling(window=7).mean()
    return df


calculate_moving_average(df5)
```

```python
#11 
import pandas as pd
import calendar
def add_weekly_column(df):
    df['Date'] = pd.to_datetime(df['Date'])
    df['Weekday']= df['Date'].apply(lambda x: calendar.day_name[x.weekday()])
    return df


df = pd.DataFrame({'Date': ['2024-02-20','2024-02-21','2024-02-22','2024-02-23','2024-02-24','2024-02-25']})
add_weekly_column(df)
```

```python
#12

def select_dates_in_range(df):
    start_date = pd.Timestamp('2023-01-01')
    end_date = pd.Timestamp('2023-01-31')
    return df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]



df = pd.DataFrame({'Date': pd.date_range(start='2023-01-01', end='2023-02-01')})
df = select_dates_in_range(df)
print(df)
```
