### Q1: Explain the following with an example:
1. **Artificial Intelligence (AI)**: 
   AI refers to the simulation of human intelligence in machines that are designed to think and act like humans. Example: A chatbot that can answer customer service queries.
   
2. **Machine Learning (ML)**: 
   ML is a subset of AI that involves the use of algorithms and statistical models to enable computers to improve at tasks with experience. Example: Email spam filtering.
   
3. **Deep Learning (DL)**: 
   DL is a subset of ML that uses neural networks with many layers to analyze various factors of data. Example: Image recognition systems.

### Q2: What is supervised learning? List some examples of supervised learning.
Supervised learning is a type of machine learning where the model is trained on labeled data. The algorithm learns from the input-output pairs and makes predictions based on the patterns identified.

**Examples:**
- Linear Regression
- Support Vector Machines (SVM)
- Decision Trees
- Neural Networks

### Q3: What is unsupervised learning? List some examples of unsupervised learning.
Unsupervised learning is a type of machine learning that deals with unlabeled data. The algorithm tries to learn the patterns and structure from the data without any specific guidance on what the output should look like.

**Examples:**
- K-means Clustering
- Hierarchical Clustering
- Principal Component Analysis (PCA)
- Association Rules

### Q4: What is the difference between AI, ML, DL, and DS?
- **AI (Artificial Intelligence)**: The broadest concept, referring to machines designed to act intelligently.
- **ML (Machine Learning)**: A subset of AI that involves machines learning from data.
- **DL (Deep Learning)**: A subset of ML that uses complex neural networks.
- **DS (Data Science)**: An interdisciplinary field focused on extracting knowledge and insights from data using various techniques, including ML and statistical analysis.

### Q5: What are the main differences between supervised, unsupervised, and semi-supervised learning?
- **Supervised Learning**: Uses labeled data to train the model.
- **Unsupervised Learning**: Uses unlabeled data to find hidden patterns.
- **Semi-Supervised Learning**: Uses a mix of labeled and unlabeled data, typically with a small amount of labeled data and a large amount of unlabeled data.

### Q6: What is train, test, and validation split? Explain the importance of each term.
- **Train Split**: The portion of data used to train the model.
- **Test Split**: The portion of data used to evaluate the model's performance.
- **Validation Split**: The portion of data used to tune the model's parameters and prevent overfitting. 

These splits ensure that the model is tested on unseen data, providing a good estimate of its performance in real-world scenarios.

### Q7: How can unsupervised learning be used in anomaly detection?
Unsupervised learning can identify patterns and groupings in data. In anomaly detection, it can help to identify outliers or unusual data points that do not fit the established patterns, indicating potential anomalies.

**Supervised Learning Algorithms:**
- Linear Regression
- Logistic Regression
- Support Vector Machines (SVM)
- Decision Trees
- Random Forest
- Neural Networks

**Unsupervised Learning Algorithms:**
