```python
#1  mongodb: It stores data in flexible, JSON-like documents instead of structured tables. Each document can have different fields and structures, allowing for diverse data types and relationships.

#Non-Relational Databases (NoSQL): Discard the fixed table structure and rigid schema.
#                     Flexibility: Embrace diverse data models (documents, key-value pairs, graphs) to accommodate unstructured or semi-structured data.
#                     Scalability: Designed for horizontal scaling and handling massive data volumes.

#When to Use MongoDB Over SQL Databases:
#Unstructured or Semi-Structured Data: MongoDB excels when data doesn't fit neatly into predefined tables or has varying structures.
#Flexibility in Data Modeling: It allows for rapid changes and evolution of data structures without complex schema migrations.
```

```python
#2 features
#1. Document-Oriented Storage: Stores data in flexible, JSON-like documents, making it ideal for semi-structured or unstructured data.
#2. Schema-Free: No predefined schema required, allowing for rapid changes and evolution of data structures.
#3  Horizontal Scalability: Designed to scale horizontally by adding more servers (nodes) to handle increasing data and workloads.
#4  High Performance: nOptimized for fast reads and writes, making it suitable for real-time applications and big data analytics.
```
