```python
#2
import scipy.stats as stats


observed = [40, 35, 45, 20, 15, 45]  

expected_distribution = [0.20, 0.20, 0.20, 0.10, 0.10, 0.20]
expected = [round(p * sum(observed)) for p in expected_distribution]


chi2, p, dof, expected = stats.chi2_contingency([observed, expected])

print("Chi-square statistic:", chi2)
print("p-value:", p)
print("Degrees of freedom:", dof)


alpha = 0.05
if p < alpha:
    print("Reject the null hypothesis: The observed distribution does not match the expected distribution.")
else:
    print("Fail to reject the null hypothesis: The observed distribution matches the expected distribution.")
```

```python
#4

import math
from scipy.stats import norm

# Sample size
n = 500

# Number of smokers
x = 60

# Sample proportion
p_hat = x / n

# Standard error
se = math.sqrt(p_hat * (1 - p_hat) / n)

# Z-score for 95% confidence interval
z = norm.ppf(0.975)  # 1.96 for 95% CI

# Margin of error
margin_of_error = z * se

# Confidence interval
lower_bound = p_hat - margin_of_error
upper_bound = p_hat + margin_of_error

print("95% Confidence Interval for the true proportion of smokers:")
print(f"({lower_bound:.4f}, {upper_bound:.4f})")
```

```python
#5
import math
from scipy.stats import t

# Sample mean
sample_mean = 75

# Sample standard deviation
sample_std_dev = 12

# Sample size (assuming it's known)
sample_size = 36  # Replace with the actual sample size

# Degrees of freedom
df = sample_size - 1

# T-score for 90% confidence interval
t_score = t.ppf(0.95, df)

# Standard error
std_error = sample_std_dev / math.sqrt(sample_size)

# Margin of error
margin_of_error = t_score * std_error

# Confidence interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

print("90% Confidence Interval for the population mean:")
print(f"({lower_bound:.2f}, {upper_bound:.2f})")
```

```python
#6
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

# Degrees of freedom
df = 10

# Chi-square statistic
chi2_stat = 15

# Create a range of x values
x = np.linspace(0, 30, 100)

# Calculate the chi-square distribution
y = chi2.pdf(x, df)

# Create the plot
plt.plot(x, y, label=f"Chi-Square Distribution (df={df})")

# Shade the area corresponding to the chi-square statistic
x_shade = np.linspace(chi2_stat, 30, 100)
y_shade = chi2.pdf(x_shade, df)
plt.fill_between(x_shade, y_shade, alpha=0.3, label=f"P(X² > {chi2_stat:.2f})")

# Label the axes
plt.xlabel("Chi-Square Statistic")
plt.ylabel("Probability Density")

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

```python
#7
import math
from scipy.stats import norm

# Sample size
n = 1000

# Number of people who prefer Coke
x = 520

# Sample proportion
p_hat = x / n

# Standard error
std_err = math.sqrt(p_hat * (1 - p_hat) / n)

# Z-score for 99% confidence interval
z_score = norm.ppf(0.995)

# Margin of error
margin_err = z_score * std_err

# Confidence interval
lower_bound = p_hat - margin_err
upper_bound = p_hat + margin_err

print("99% Confidence Interval for the true proportion of people who prefer Coke:")
print(f"({lower_bound:.3f}, {upper_bound:.3f})")
```

```python
#8
import scipy.stats as stats

# Observed frequencies
observed = [55, 45]  # Heads, Tails

# Expected frequencies under the null hypothesis (fair coin)
expected = [50, 50]  # Heads, Tails

# Calculate the chi-square statistic and p-value
chi2, p, dof, expected = stats.chi2_contingency([observed, expected])

print("Chi-Square Goodness of Fit Test:")
print(f"Chi-Square Statistic: {chi2:.2f}")
print(f"P-Value: {p:.4f}")
print(f"Degrees of Freedom: {dof}")

# Determine the outcome of the test
alpha = 0.05
if p < alpha:
    print("Reject the null hypothesis: The coin is not fair.")
else:
    print("Fail to reject the null hypothesis: The coin is fair.")
```

```python
#9
import scipy.stats as stats

# Contingency table
table = [[60, 30], [140, 170]]  # Smoker/Non-Smoker, Lung Cancer/No Lung Cancer

# Calculate the chi-square statistic and p-value
chi2, p, dof, expected = stats.chi2_contingency(table)

print("Chi-Square Test for Independence:")
print(f"Chi-Square Statistic: {chi2:.2f}")
print(f"P-Value: {p:.4f}")
print(f"Degrees of Freedom: {dof}")

# Determine the outcome of the test
alpha = 0.05
if p < alpha:
    print("Reject the null hypothesis: There is a significant association between smoking status and lung cancer diagnosis.")
else:
    print("Fail to reject the null hypothesis: There is no significant association between smoking status and lung cancer diagnosis.")
```

```python
#11
import scipy.stats as stats

# Sample statistics
x_bar = 72
s = 10
n = 30

# Population parameter
mu = 70

# Calculate the t-statistic
t = (x_bar - mu) / (s / np.sqrt(n))

print("t-statistic:", t)

# Determine the degrees of freedom
df = n - 1

print("Degrees of Freedom:", df)

# Find the critical value
alpha = 0.05
critical_value = stats.t.ppf(1 - alpha/2, df)

print("Critical Value:", critical_value)

# Make a decision
if abs(t) > critical_value:
    print("Reject the null hypothesis: The population mean is significantly different from 70.")
else:
    print("Fail to reject the null hypothesis: The population mean is not significantly different from 70.")
```
